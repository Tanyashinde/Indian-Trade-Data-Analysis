## Now lets try to solve our first Question : 
## 1.What did India export the most in any given year?

```python
df_18= df[df["year"]==2018]
df_18
```

```python
df_18[df_18["Commodity"]=="MEAT AND EDIBLE MEAT OFFAL."]
```

```python
df_18.sort_values('value',inplace=True,ascending=False)
df_18
```

```python
Top_10_trade= df_18[:10]
Top_10_trade
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
plt.title('Top 10 Indian Trades from the year 2018')

values= Top_10_trade["value"]
countries= Top_10_trade["country"]
ax.pie(values,labels=countries,autopct='%1.2f%%')
plt.show()
```

# 2. Which year had the most Exports?

```python
df.head(20)
```

```python
most_trade= df.groupby("year").sum()
most_trade
```

```python
most_trade.sort_values('value',inplace=True,ascending=False)
most_trade
```

```python
# most_trade["value"].plot.hist();

most_trade["value"].plot();
plt.title('Treade values of each year from 2010-2018')
plt.xlabel('Year')
plt.ylabel('Trade Values')
plt.show()
```

# 3.Which commodity forms a major chunk of trade?

```python
df.head(20)
```

```python
com_trade= df.groupby(['Commodity'])['value'].sum().reset_index()
```

```python
com_trade
```

```python
com_trade.sort_values('value',inplace=True,ascending=False)
```

```python
# fig,ax= plt.subplots()
# ax.barh(com_trade['Commodity'],com_trade['value'])

plt.figure(figsize=(100,100))
plt.title('Commodity trade values exported from 2010-1018')
plt.barh(com_trade['Commodity'],com_trade['value'], align='center')
plt.xlabel('Value in (million $)')
plt.ylabel('commodities')
plt.show()
```

```python
top_10_trades= com_trade[:10]
top_10_trades
```

```python
plt.figure(figsize=(20,20))
plt.title('Top 10 commodity trade values from 2010-2018')
plt.barh(top_10_trades['Commodity'],top_10_trades['value'], align='center')
plt.xlabel('Value in (million $)')
plt.ylabel('commodities')
plt.show()
```

# 4.How has the trade between India and any given country grown over time? 

```python
country_trade= df[df['country']=="RUSSIA"]
country_trade
```

```python
trade_over_time=country_trade.groupby(['year'])['value'].sum().reset_index()
trade_over_time
```

```python
trade_over_time.sort_values('value',inplace=True,ascending=True)
```

```python
trade_over_time
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
plt.title('trade values of each year from 2010-2018 between India and Russia')
ax.pie(trade_over_time['value'],labels=trade_over_time['year'],autopct='%1.2f%%')
plt.show()
```

```python
trade_over_time.plot(x="year",y="value",kind="bar");
```
